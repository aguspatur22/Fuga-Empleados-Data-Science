#### EDA VARIABLES CATEGÓRICAS

Este código genera gráficos de barras horizontales para visualizar la distribución de las categorías en las variables categóricas. Cada gráfico representa una variable categórica y muestra la frecuencia de cada categoría. Esto permite entender mejor la composición de los datos categóricos y detectar posibles desequilibrios o patrones en ellos.

```python
def graficos_eda_categoricos(cat):

    #Calculamos el número de filas que necesitamos
    from math import ceil
    filas = ceil(cat.shape[1] / 2)

    #Definimos el gráfico
    f, ax = plt.subplots(nrows = filas, ncols = 2, figsize = (16, filas * 6))

    #Aplanamos para iterar por el gráfico como si fuera de 1 dimensión en lugar de 2
    ax = ax.flat

    #Creamos el bucle que va añadiendo gráficos
    for cada, variable in enumerate(cat):
        cat[variable].value_counts().plot.barh(ax = ax[cada])
        ax[cada].set_title(variable, fontsize = 12, fontweight = "bold")
        ax[cada].tick_params(labelsize = 12)
```

```python
graficos_eda_categoricos(df.select_dtypes('O')) #Porque las categorias son del tipo "object"
```

Viendo los graficos generados podemos detectar que:

* La variable mayor_edad siempre tiene un unico valor, por lo que es una CONSTANTE y no tiene sentido su analisis, asi que la eliminamos
* Para imputar las variables debemos elegir el valor que minimice el error, es decir, imputar por el valor mayoritario:
  * educacion: imputar por 'Universitaria'
  * satisfaccion_trabajo: imputar por 'Alta'
  * implicacion: imputar por 'Alta'

```python
df.drop(columns = 'mayor_edad', inplace = True)

df['educacion'] = df['educacion'].fillna('Universitaria')

df['satisfaccion_trabajo'] = df['satisfaccion_trabajo'].fillna('Alta')

df['implicacion'] = df['implicacion'].fillna('Alta')
```

#### EDA VARIABLES NUMÉRICAS

Este código calcula estadísticas descriptivas, como la media, la mediana y otros valores importantes, para todas las variables numéricas en el conjunto de datos. Estas estadísticas proporcionan información sobre la distribución y tendencia central de los datos.
Esto es útil en esta etapa de análisis de datos previos para comprender la naturaleza de las variables numéricas y detectar posibles valores atípicos o patrones en los datos

```python
def estadisticos_cont(num):
    #Calculamos describe
    estadisticos = num.describe().T
    #Añadimos la mediana
    estadisticos['median'] = num.median()
    #Reordenamos para que la mediana esté al lado de la media
    estadisticos = estadisticos.iloc[:,[0,1,8,2,3,4,5,6,7]]
    #Lo devolvemos
    return(estadisticos)
```

```python
estadisticos_cont(df.select_dtypes('number'))
```

Viendo los resultados concluimos que:

* Empleados y horas quincena solo tienen un valor, asi que como en el mismo caso de la variable mayor_edad lo mas conveniente es eliminarla
* Sexo tiene 4 valores, por lo que claramente hay un error con los datos. Se podria consultar al area de sistemas sobre esto, o como haremos en este caso, eliminarla.
Recordemos que en el analisis de nulos determinamos que habia que imputar la variable sexo, pero como en este paso decidimos eliminarla entonces termino ese analisis


```python
df.drop(columns = ['empleados','sexo','horas_quincena'], inplace = True)
df
```

### ¿Cual es el impacto económico de este problema?

Según el estudio "Cost of Turnover" del Center for American Progress:

* El coste de la fuga de los empleados que ganan menos de 30000 es del 16,1% de su salario

* El coste de la fuga de los empleados que ganan entre 30000-50000 es del 19,7% de su salario

* El coste de la fuga de los empleados que ganan entre 50000-75000 es del 20,4% de su salario

* El coste de la fuga de los empleados que ganan más de 75000 es del 21% de su salario

```python
# Creamos una nueva variable salario_ano del empleado
df['salario_ano'] = df.salario_mes.transform(lambda x: x*12)
df[['salario_mes','salario_ano']]
```

```python
# Calculamos el impacto económico de cada empleado si deja la empresa

#Lista de condiciones
condiciones = [(df['salario_ano'] <= 30000),
               (df['salario_ano'] > 30000) & (df['salario_ano'] <= 50000),
               (df['salario_ano'] > 50000) & (df['salario_ano'] <= 75000),
               (df['salario_ano'] > 75000)]

#Lista de resultados
resultados = [df.salario_ano * 0.161, df.salario_ano * 0.197, df.salario_ano * 0.204, df.salario_ano * 0.21]

#Aplicamos select
df['impacto_abandono'] = np.select(condiciones,resultados, default = -999)

df
```

## MODELO DE MACHINE LEARNING

Cuando trabajamos con modelos de ML debemos hacer una copia de los datos ya que los podemos modificar, y lo ideal es seguir teniendo el dataframe original.

Respecto a la preparacion de los datos, hay ciertas restricciones:
* obligatorias: no puede haber nulos y todas las variables deben estar en formato numerico
* opcionales: (para mejorar la capacidad predictiva) podriamos reescalar variables (poner en misma escala 0-1, 0-100, etc)

```python
df_ml = df.copy()
```

```python
df_ml.info()
```

### PREPARACIÓN DE LOS DATOS PARA LA MODELIZACIÓN

#### Transformar todas las variables categóricas a númericas

La técnica de One-Hot Encoding es una forma de convertir variables categóricas en variables numéricas. Imaginate que tenes una variable categórica como "color" con tres categorías: rojo, azul y verde.

Con One-Hot Encoding, en lugar de representar cada categoría con un solo número (como 1 para rojo, 2 para azul y 3 para verde), creamos una nueva columna para cada categoría. En estas nuevas columnas, asignamos un 1 si la observación pertenece a esa categoría y un 0 si no.

Por ejemplo, si un registro tiene el color rojo, la nueva columna "rojo" tendría un valor de 1 y las columnas "azul" y "verde" tendrían un valor de 0.

Esta técnica es útil porque permite que los algoritmos de aprendizaje automático interpreten correctamente las variables categóricas, ya que algunos algoritmos pueden tener dificultades con las variables que no son numéricas. Además, ayuda a evitar que el modelo asuma que hay algún tipo de relación ordenada entre las categorías, lo que podría no ser cierto en todos los casos.

```python
from sklearn.preprocessing import OneHotEncoder

#Categóricas
cat = df_ml.select_dtypes('O')

#Instanciamos
ohe = OneHotEncoder(sparse = False)

#Entrenamos
ohe.fit(cat)

#Aplicamos
cat_ohe = ohe.transform(cat)

#Ponemos los nombres
cat_ohe = pd.DataFrame(cat_ohe, columns = ohe.get_feature_names_out(input_features = cat.columns)).reset_index(drop = True)
```

```python
cat_ohe
```

#### Dataframe final

Seleccionamos las variables numéricas para poder juntarlas a las cat_hoe

```python
num = df.select_dtypes('number').reset_index(drop = True)
```

Las juntamos todas en el dataframe final

```python
df_ml = pd.concat([cat_ohe,num], axis = 1)
df_ml
```

### DISEÑO DE LA MODELIZACIÓN

#### Separación predictoras y target

```python
x = df_ml.drop(columns='abandono')
y = df_ml['abandono']
```

#### Separación train y test

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.3)
```

### ENTRENAMIENTO DEL MODELO SOBRE TRAIN

***¿Por que elegimos este algoritmo en especial?***

En el mundo de los algoritmos de Machine Learning (ML) tenés distintas opciones, cada una con sus características y usos particulares. La idea detrás de todos estos algoritmos es tratar de sacar información de las variables que tenés para predecir algo, lo que llamamos la "target". Pero cada algoritmo se lleva mejor con ciertos tipos de datos y situaciones.

Por ejemplo, tenés tres grandes familias de algoritmos:

* Regresiones: Imaginate que tenés un montón de datos y querés ver cómo una variable afecta a otra. Acá lo que se hace es ajustar una línea o un plano (dependiendo de si tenés una o más variables) para tratar de predecir la variable objetivo. En criollo, es como cuando tratás de encontrar la relación entre el precio de una casa y su tamaño.

* Redes neuronales: Estos son como el cerebro artificial de la computadora. Se basan en el funcionamiento de las neuronas en nuestro cerebro y son muy útiles para problemas muy complejos donde hay patrones difíciles de detectar para nosotros, los humanos.

* Árboles de decisión: Este es un poco más simple. Imaginate que tenés un montón de reglas y vas decidiendo en base a esas reglas qué camino tomar. Por ejemplo, si querés predecir si alguien va a comprar o no un producto, podés tener reglas como "si la persona tiene más de 30 años y gana más de $50.000 al mes, entonces es probable que compre".

El árbol de decisión funciona así: arranca con todos los casos que tenés y va viendo qué cumplen con ciertas condiciones y cuáles no. Y va tomando decisiones así, de a pasitos, hasta llegar a una predicción.

Una cosa importante del árbol de decisión es que es súper explicativo. O sea, te muestra bien por qué llegó a determinada conclusión. Por eso, en el caso nuestro, que queremos entender por qué se van los empleados de la empresa, viene bárbaro. Pero si estás en un banco y necesitás detectar transacciones fraudulentas rápido, ahí no te conviene, porque necesitás velocidad más que explicaciones.

```python
from sklearn.tree import DecisionTreeClassifier

#Instanciar
ac = DecisionTreeClassifier(max_depth=4)

#Entrenar
ac.fit(train_x,train_y)
```

### PREDICCIÓN Y VALIDACIÓN SOBRE TEST

```python
# Predicción
pred = ac.predict_proba(test_x)[:, 1]
pred[:20]
```

```python
# Evaluación
from sklearn.metrics import roc_auc_score

roc_auc_score(test_y,pred)
```

***¿Como evaluamos que nuestro modelo funciona correctamente?***

Mediante una métrica específica llamada "área bajo la curva ROC" (ROC AUC, por sus siglas en inglés). Lo que hace el roc_auc_score es comparar las predicciones que hizo tu modelo con las respuestas reales que tenés en tus datos de prueba. Y te da un número entre 0 y 1, donde 1 significa que tu modelo está acertando todas las predicciones perfectamente, y 0.5 significa que está tirando monedas al aire y no es mejor que adivinar al azar. Entonces, cuanto más cerca esté del 1, mejor está funcionando tu modelo.

* menor a 0.7 no es bueno
* entre 0.7 y 0.8 es practico
* mayor a 0.8 es muy bueno
